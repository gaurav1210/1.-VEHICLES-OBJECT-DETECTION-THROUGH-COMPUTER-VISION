Question 1: Taylor Series

Take the taylor series expansion for $f(x)=e^x$ as,
$f(x) = e^{x} \approx \sum_{n=0}^{\infty} \frac{x^n}{n!}$.


Compute the following.


(1) Value of e for n = 5 approximation.


(2) Value of 1/e upto 10th decimal place. (Exercise)

````python
# Answer 1a
n = 5 # Given n
x = 1 # e^1 = e
e_val = 1 # First term
fact =1 # Initialise with 0!
for i in range(1,n):
    fact = fact*i
    e_val = e_val + x**i/fact
print(e_val,np.e)
````

````python
n = 5  # Given n
x = 1  # e^1 = e
e_val = 1  # First term
fact = 1  # Initialise with 0!
for i in range(1, n + 1):
    fact = fact * i
    e_val = e_val + x ** i / fact
print(f"Value of e for n = {n} approximation: {e_val}")

# Part 2: Value of 1/e up to 10th decimal place
x = -1
e_inverse_val = 1
fact = 1
for i in range(1, n + 1):
    fact = fact * i
    e_inverse_val = e_inverse_val + x ** i / fact
print(f"Value of 1/e for n = {n} approximation: {e_inverse_val:.10f}")
````

Question 3: Lagrange Polynomial Approximation (Exercise)

(a) Compute langrage polynomial for $f(x) = x**2$ and at x = 3.14159, if known points are located at $x = {0,0.1,0.45,0.8}$.

(b) Plot the approximated function and the true function $f(x)$.

(c) Calculate the truncation error in the approximation against the true value.

````python
# prompt: Question 3: Lagrange Polynomial Approximation (Exercise)
# (a) Compute langrage polynomial for  𝑓(𝑥)=𝑥∗∗2  and at x = 3.14159, if known points are located at  𝑥=0,0.1,0.45,0.8 .
# (b) Plot the approximated function and the true function  𝑓(𝑥) .
# (c) Calculate the truncation error in the approximation against the true value.

import numpy as np
import matplotlib.pyplot as plt

def lagrange_polynomial(x, x_points, y_points):
    n = len(x_points)
    y = 0
    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if i != j:
                term = term * (x - x_points[j]) / (x_points[i] - x_points[j])
        y = y + term
    return y

# (a) Compute Lagrange polynomial for f(x) = x^2 at x = 3.14159
x_points = np.array([0, 0.1, 0.45, 0.8])
y_points = x_points**2
x = 3.14159
approximated_value = lagrange_polynomial(x, x_points, y_points)
true_value = x**2
print(f"Approximated value at x = {x}: {approximated_value}")
print(f"True value at x = {x}: {true_value}")

# (b) Plot the approximated function and the true function f(x)
x_vals = np.linspace(0, 1, 100)  # Adjust the range as needed
y_true = x_vals**2
y_approx = [lagrange_polynomial(val, x_points, y_points) for val in x_vals]

plt.plot(x_vals, y_true, label='True Function (f(x) = x^2)')
plt.plot(x_vals, y_approx, label='Lagrange Polynomial Approximation')
plt.scatter(x_points, y_points, color='red', label='Known Points')  # Plot known points
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Lagrange Polynomial Approximation of f(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()


# (c) Calculate the truncation error
truncation_error = abs(approximated_value - true_value)
print(f"Truncation error at x = {x}: {truncation_error}")
````

Question 4: Truncation Error vs Round-off Error

Take the taylor series expansion for $f(x)=e^x$ as,
$f(x) = e^{x} \approx \sum_{n=0}^{\infty} \frac{x^n}{n!}$.

(a) Compute round off errors to nearest number for the value of e, at 1,5,8,10,15,20 decimal places.$\\$
(b) Compute the truncation errors for first 10, 15, 200, 500, 1000 terms.

Use:
```python
from mpmath import mp
def get_round_e_value(n):
   '''Input n is the number of decimal
   digits you want e round-off to nearest number'''
   mp.dps=n+1
   return(mp.e)
```

````python
from mpmath import mp
def get_round_e_value(n):
   '''Input n is the number of decimal
   digits you want e round-off to nearest number'''
   mp.dps=n+1
   return(mp.e)

trun_num = np.array([10,15,200,500,1000])
trun_array = np.zeros(len(trun_num))
x = 1 # e^1 = e
for j in range(len(trun_num)):
   e_val = 1 # First term
   fact =1 # Initialise with 0!
   for i in range(1,trun_num[j]):
      fact = fact*i
      e_val = e_val + x**i/fact
   trun_array[j] = abs(e_val-np.e)
   print(trun_array[j])

plt.plot(trun_num,trun_array)
plt.yscale('log')
plt.show()

roff_num = np.array([1,5,8,10,15,20])
print(get_round_e_value(roff_num[0]))
rd_arr = np.zeros(len(roff_num))

for i in range(len(roff_num)):
   e_value= get_round_e_value(roff_num[i])
   rd_arr[i] = abs(np.e-e_value)
   print(rd_arr[i])

plt.plot(roff_num,rd_arr)
plt.yscale('log')
plt.show()
````
